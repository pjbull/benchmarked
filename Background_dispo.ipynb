```python
%matplotlib inline
import json
import fnmatch
import numpy as np
import networkx as nx
import requests
from pattern import web
import re
import matplotlib.pyplot as plt
from itertools import combinations 
import operator
import pandas as pd
import numpy as np
df = pd.read_csv('data/SCDB_2013_01_justiceCentered_Citation_Cleaned.csv')

def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecessary plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
        
df_filtered = df[df.columns[25:28]]
id_col = df[df.columns[0]]
df_filtered['caseId'] = id_col
df_caseId = df_filtered.groupby('caseId')
print df_filtered.columns
reason_for_cert = []
for a, b in df_caseId:
    if not np.isnan(b['certReason'].unique()):
        reason_for_cert.append((a, b['certReason'].unique()))
print len(reason_for_cert)
```

```python
URL = 'http://scdb.wustl.edu/documentation.php?var=certReason'
xml = requests.get(URL).text.encode('ascii', 'ignore')
dom = web.DOM(xml)
body = dom.body
states = []

data_list = []
for table in body.by_tag('table'):
    if len(table) >20:
        for data in table.by_tag('tr'):
            for item in data.by_tag('td'):
                a =  item.content
                a = a.replace('\t', '')
                a = str(a.replace('\n', ''))
                data_list.append(a)
count = 0
number_to_cert_reason = {}
for item in data_list:
    if count%2 == 0:
        number = item
    if count%2 != 0:
        reason = item
        number_to_cert_reason[number] = reason
    count += 1
    
print number_to_cert_reason
                
```

```python
certreas_counts = {}
keys = number_to_cert_reason.keys()
for i in range(len(number_to_cert_reason.keys())):
    count = str(keys[i])
    certreas_counts[count] = 0

for item in reason_for_cert:
    num = str(int(item[1][0]))
    try:
        certreas_counts[num] += 1
    except KeyError:
        pass

names = []
counts = []
names_counts = {}
for reas_num in number_to_cert_reason.keys():
    names.append(number_to_cert_reason[reas_num])
    counts.append(certreas_counts[reas_num])
    names_counts[number_to_cert_reason[reas_num]] = certreas_counts[reas_num]
                
    

y_pos = np.arange(len(number_to_cert_reason.keys()))
fig = plt.figure()
plt.barh(y_pos, counts, align='center')
plt.yticks(y_pos,names)
fig.set_size_inches(5, .2*len(number_to_cert_reason.keys()))
plt.ylim([-1, len(number_to_cert_reason.keys())])
plt.xlabel('Absolute Case Count')
plt.title('Reasons for Cert')
remove_border()
plt.show()
```
